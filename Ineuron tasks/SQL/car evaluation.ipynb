```python
cursor.execute('create table car(col1 VARCHAR(10),col2 VARCHAR(10),col3 VARCHAR(10),col4 VARCHAR(10),col5 VARCHAR(10),col6 VARCHAR(10),col7 VARCHAR(10))')
```

```python
import csv
with open('car.data','r') as f:
    car_data = csv.reader(f,delimiter='\n')
    query = """insert into car (col1,col2,col3,col4,col5,col6,col7) values(%s,%s,%s,%s,%s,%s,%s)"""
    for i in car_data:
        split_data = i[0].split(',')
        cursor.execute(query,split_data)
        
```
